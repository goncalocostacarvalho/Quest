```python
def full_game():
    import time
    def act_break(): #function to be used to separate the different chapters
        print("")
        print("")
    def print_slow(text, delay=1.5): #function to be used to delay the next message to user be able to read the message
        print(text)
        time.sleep(delay)
    def scenario():
        print_slow("In a mystical land, there exists a prophecy about an Enchanted Orb that has the power to bring balance to the realms.\n")
        print_slow("The orb, however, has been lost for centuries.\n")
        print_slow("Legends speak of a hidden cave where the orb is said to be guarded by magical creatures and traps.\n")
        valid_answer=False
        while not valid_answer:
            try:
                response=input("Are you ready to embrace the adventure?\nType yes or no.\n")
                if response.lower() == "yes":
                    print_slow("\nLet's get started!")
                    valid_answer=True
                elif response.lower()== "no":
                    print_slow("Sorry but you are brave and need to embrace the adventure!\n")
                else:
                    raise ValueError("Invalid answer! Type yes or no.\n")
            except ValueError as error:
                print_slow(f"Error: {error}\n")
    def choose_character():
        characters=["Warrior", "Wise", "Enchantress", "Guardian" ]
        name=input("What is your name? ")
        valid_answer=False
        while not valid_answer:
            try:
                character=input(f"Guess your character from {characters}.\n")
                if character.lower()=="warrior":
                    print_slow(f"\nCongrats! Now, you are {name}, the {character}. Let the adventure begin!")
                    valid_answer=True
                else:
                    raise ValueError("Invalid Answer! To start this game you need to choose another character.\n")
            except ValueError as error:
                print_slow(f"\nError: {error}")    
    def act_1():
        print_slow(f"You are at home and suddenly you hear knockning your door.\n")
        valid_answer=False
        while not valid_answer:
            try:
                action_1=input("What you do? Type nothing or check.\n")
                if action_1.lower()=="check":
                    print_slow(f"\nYou found no one at the door. But on the ground there is a letter.\n")
                    print_slow(f"This letter instructs you to meet Eldor the Wise in Eldoria.\n")
                    print_slow("Now you have another decision to make!\n")
                    valid_answer=True 
                elif action_1.lower()=="nothing":
                    print_slow(f"You are not the type of people who does nothing. Try to check what happened.\n")
                else:
                    raise ValueError("Invalid answer. Type nothing or check.")
            except ValueError as error:
                print_slow(f"Error: {error}\n")  
        valid_answer_2=False
        while not valid_answer_2:
            try:
                action_2=input("You either visit Eldor the Wise or stay at home.\nType visit or stay.\n")
                if action_2.lower()=="visit":
                    print_slow(f"\nYou arrive Eldoria and meet Eldor the Wise.")
                    print_slow(f"\nHe shares the prophecy with you and informs you of the hidden cave's general location.")
                    print_slow(f"\nNow you are in charge to find the Enchanted Orb and to bring peace and balance to the realms.")
                    valid_answer_2=True
                elif action_2.lower()=="stay":
                    print_slow(f"\nIf you decide to stay at home, there will be no peace and balance! Please rethink your action.")
                else:
                    raise ValueError("Invalid answer. Type visit or stay.")    
            except ValueError as error:
                print_slow(f"\nError: {error}")
    def act_2():
        print_slow(f"With the hidden cave's general location in mind, you decide to start your adventure.\n")
        print_slow(f"On your way, you find Sylvan, the Enchantress who for a long time didn't see anyone cross her mystical forest.\n")
        print_slow(f"She starts to talk with you and soon you understand that you will need her help to access the hidden cave.\n")
        print_slow(f"She tells you she can give her help but you need to find a rare flower.\n")
        valid_answer=False
        while not valid_answer:
            try:
                action_3=input("What you decide to do?\nType explore or chat.\n")
                if action_3.lower()=="chat":
                    print_slow(f"\nYou decide to chat a bit more!")
                    print_slow(f"\nYou get to know each other better but she insists that only with the rare flower can guide you.\n")
                elif action_3.lower()=="explore":
                    print_slow(f"\nYou decide to explore!")
                    print_slow(f"\nWhile you explore the mystical forest, you discover an hidden grove and the rare flower Sylvan wants.")
                    print_slow(f"\nYou return and give the flower to Sylvan.")
                    valid_answer=True
                else:
                    raise ValueError("Invalid answer. Type explore or chat.")
            except ValueError as error:
                print_slow(f"\nError: {error}") 
        print_slow(f"\nSylvan decides to give her guidance and takes you to the cave.")
    def act_3():
        tasks=["puzzles", "traps", "creatures"]
        print_slow(f"Armed with Sylvan's guidance, you enter the cave.\n")
        print_slow(f"Looking around you see puzzles, traps and creatures guarding the path.\n")
        superpowers=set()
        valid_answer=False
        while not valid_answer:
            task=input(f"Choose a task you want to complete from {tasks}.\nIn case you don't want to complete any task, type pass.\n")
            try: 
                if task.lower()=="puzzles":
                    print_slow(f"\nCongrats! After hours of patience, the task was complete. You obtained patience as a super power!")
                    finish=input("\nDo you want to complete more tasks?\nType yes or no.\n")
                    superpowers.add("patience")
                    valid_answer=finish=="no"
                elif task.lower()=="traps":
                    print_slow(f"\nCongrats! You were able to avoid all the traps. You obtained intelligence as a super power!")
                    finish=input("\nDo you want to complete more tasks?\nType yes or no.\n")
                    superpowers.add("intelligence")
                    valid_answer=finish=="no"
                elif task.lower()=="creatures":
                    print_slow(f"\nCongrats! After hours of fight, the task was complete. You obtained strenght as a super power!")
                    finish=input("\nDo you want to complete more tasks?\nType yes or no.\n")
                    superpowers.add("strength")
                    valid_answer=finish=="no"
                elif task.lower()=="pass":
                    print_slow(f"\nCongrats! You obtained no super power!")
                    valid_answer=True
                else:
                    raise ValueError("Invalid answer. Type one of the tasks or pass.")
            except ValueError as error:
                print_slow(f"\nError: {error}")
        print_slow(f"\nYour choices inside the cave influenced your abilities.\nHope you are ready for your biggest challenge in this journey.")
        print(f"\nYou accquired the following superpowers: {superpowers}.")
        return superpowers
    def act_4(superpowers):
        approach=["outsmart", "battle", "debate"]
        print_slow(f"After facing some challenges or not, in the heart of the cave you found Grimclaw, the Guardian.")
        valid_answer=False
        while not valid_answer:
            if len(superpowers)==0:
                print_slow(f"\nYou didn't face any challenge and don't have any superpower.\nYou can not pass through Grimclaw.")
                print_slow(f"\nYour journey ends here.")
                return False
            else:
                print_slow(f"\nTo reach the Enchanted Orb, you can face Grimclaw in different ways: {approach}.")
                decision=input("\nChoose the way you want to face Grimclaw.\n")
            try:
                if decision.lower()=="outsmart": 
                    if "intelligence" in superpowers:
                        print_slow(f"\nDue the intelligence you gained before, you were able to outsmart and surprass Grimclaw.") 
                        valid_answer=True
                    else:
                        print_slow(f"\nYou don't have the superpower needed to outsmart Grimclaw. Choose another way to face Grimclaw.")
                elif decision.lower()=="battle":
                    if "strength" in superpowers:
                        print_slow(f"\nDue the strength you gained before, you enter on an intense battle with Grimclaw and you claim the victory.") 
                        valid_answer=True
                    else:
                        print_slow(f"\nYou don't have the superpower needed to battle Grimclaw. Choose another way to face Grimclaw.")
                elif decision.lower()=="debate": 
                    if "patience" in superpowers:
                        print_slow(f"\nDue the patience you gained before, you explained to Grimclaw that the Enchanted Orb was necessary to bring peace and balance to the realms.\nYou became friends and Grimclaw let you pass.")
                        valid_answer=True
                    else:
                        print_slow(f"\nYou don't have the superpower needed to explain the situation you face to Grimclaw. Choose another way to face Grimclaw.")
                else:
                    raise ValueError("Invalid answer. Type one of the correct ways to face Grimclaw.")
            except ValueError as error:
                print_slow(f"\nError: {error}")
        print_slow(f"\nAfter this, you are almost on the end of your journey.")
        return True
    def act_5():
        print_slow(f"You finally reach the chamber containing the Enchanted Orb.\n")
        print_slow(f"Now you have a final decision to make.\nTake the orb with you or not.\n")
        print_slow(f"A lot of thoughts come to your mind, especially the one:\nYou either die a hero or live long enough to become the villain.\n")
        valid_answer=False
        while not valid_answer:
            try:
                final_decision=input("In the end what do you choose to do?\nType abandon or reclaim.\n")
                if final_decision.lower()=="abandon":
                    print_slow(f"\nYou let your invasive thougts get the most of you and you abandon the chamber.\nYour journey was impressive but you weren't able to bring peace and balance to the realms.")
                    valid_answer=True
                    return
                elif final_decision.lower()=="reclaim":
                    print_slow(f"\nYou reclaimed the enchanted orb and you were able to bring peace and balance to the realms.\nYou became a hero to your land!")
                    valid_answer=True
                else:
                    raise ValueError("Invalid answer. Type abandon or reclaim.")
            except ValueError as error:
                print_slow(f"\nError: {error}.")

    print_slow("Welcome to the Enchanted Orb Adventure!\n")
    print_slow("Your journey begins now.\n")
    scenario()
    act_break()
    choose_character()
    act_break()
    act_1()
    act_break()
    act_2()
    act_break()
    superpowers = act_3()
    act_break()
    if act_4(superpowers):
        act_break()
        act_5()
    print_slow(f"\nThe end! Hope you liked!")
```

```python
full_game()
```
